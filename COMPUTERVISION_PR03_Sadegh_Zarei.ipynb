```python
# 🔥 Advanced Vision School - Project 3: Smart Text-to-Image App
# 🌟 Author: [Sadegh Zarei]

!pip install --quiet diffusers transformers accelerate gradio

from diffusers import StableDiffusionPipeline
import torch
import gradio as gr

# 🟢 Load pipelines for both models
pipe_sd15 = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
).to("cuda")

pipe_sdxl = StableDiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    torch_dtype=torch.float16,
    variant="fp16"
).to("cuda")

# 🎨 Function to generate image
def generate_image(prompt, model_choice, cfg_scale, width, height, seed):
    if model_choice == "SD15":
        pipe = pipe_sd15
    else:
        pipe = pipe_sdxl

    generator = torch.Generator("cuda").manual_seed(seed)
    image = pipe(
        prompt,
        guidance_scale=cfg_scale,
        width=width,
        height=height,
        generator=generator
    ).images[0]
    return image

# ✨ Gradio Interface
with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown(
        """
        # 🖼️ **Smart Text-to-Image Generator**
        **Advanced Vision School - Project 3**
        """
    )
    with gr.Row():
        with gr.Column(scale=2):
            prompt = gr.Textbox(
                label="Enter your prompt",
                placeholder="e.g., A futuristic cityscape at sunset",
            )
            model_choice = gr.Radio(
                ["SD15", "SDXL"],
                label="Select Model",
                value="SD15"
            )
            cfg_scale = gr.Slider(
                1.0, 20.0, value=7.5, step=0.5,
                label="CFG Scale"
            )
            size = gr.Radio(
                ["512x512", "768x512", "768x768"],
                label="Image Size",
                value="512x512"
            )
            seed = gr.Slider(
                0, 999999, step=1, value=42,
                label="Random Seed"
            )
            generate_btn = gr.Button("Generate Image 🚀")

        with gr.Column(scale=3):
            output = gr.Image(label="Generated Image")

    def process(prompt, model_choice, cfg_scale, size, seed):
        width, height = map(int, size.split("x"))
        img = generate_image(prompt, model_choice, cfg_scale, width, height, seed)
        return img

    generate_btn.click(
        process,
        inputs=[prompt, model_choice, cfg_scale, size, seed],
        outputs=output
    )

# 🚀 Launch Gradio app
demo.launch(share=True)
```
